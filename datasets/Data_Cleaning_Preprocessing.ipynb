```python
import pandas as pd

df = pd.read_csv("Amazon Sale Report.csv")

print(df.shape)
print(df.isnull().sum())
```

```python
df.drop(columns=['Unnamed: 22'], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Date'].head(50)
```

```python
df['Month']=df['Date'].dt.month_name()
df['Month_Num']=df['Date'].dt.month
df['Quarter']=df['Date'].dt.quarter
df['Weekday']=df['Date'].dt.day_name()
df['Year']=df['Date'].dt.year
```

```python
df[['Date','Month','Month_Num','Quarter','Weekday','Year']].head()
```

```python
df[df['Amount'].isna()][['Status','Amount']].head(20)
```

```python
df[df['Amount'].isna()]['Status'].value_counts()
```

```python
missing_amount = df[df['Amount'].isna()]

missing_amount[
    missing_amount['Status'] != 'Cancelled'
][['Order ID','Status','Qty','Category']]
```

```python
df_cancelled = df[df['Status'] == 'Cancelled']
df_non_cancelled = df[df['Status'] != 'Cancelled']
```

```python
df_cancelled['Amount'] = df_cancelled['Amount'].fillna(0)
```

```python
df.loc[
    (df['Status'] == 'Cancelled') &
    (df['Amount'].isna()),
    'Amount'
] = 0
```

```python
df[
    (df['Amount'].isna()) &
    (df['Status'] != 'Cancelled')
]['Status'].value_counts()
```

```python
df[
    (df['Amount'].isna()) &
    (df['Status'] != 'Cancelled')
][['Order ID','Status','Qty','Category','currency']].head(20)
```

```python
df[
    (df['Amount'].isna()) &
    (df['currency'].isna())
].shape
```

```python
df = df[
    ~(
        (df['Amount'].isna()) &
        (df['currency'].isna()) &
        (df['Status'] != 'Cancelled')
    )
]
```
